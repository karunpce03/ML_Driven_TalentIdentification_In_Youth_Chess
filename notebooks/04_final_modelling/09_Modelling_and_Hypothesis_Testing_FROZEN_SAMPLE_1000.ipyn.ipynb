```python
# ============================================================
# Create model_df and save final modelling dataset
# ============================================================

import pandas as pd
import numpy as np
from pathlib import Path

# New project path
PROJECT_DIR = Path.home() / "Desktop" / "Arun" / "Arun" / "DBA-Walsh" / "SecondYr" / "Capstone" / "ML_Driven_TalentIdentification_In_Youth_Chess"

# Old data folder where current CSVs exist
OLD_DATA_DIR = Path.home() / "Downloads" / "fide_history"

# Project folders
PROCESSED_DIR = PROJECT_DIR / "data" / "processed"
OUTPUT_TABLES_DIR = PROJECT_DIR / "outputs" / "tables"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
OUTPUT_TABLES_DIR.mkdir(parents=True, exist_ok=True)

# Input files
PLAYER_PATH = OLD_DATA_DIR / "player_sample_1000.csv"
CALC_PATH = OLD_DATA_DIR / "fide_calculations_all_players_standard_clean_combined.csv"

# Load files
players = pd.read_csv(PLAYER_PATH)
calc = pd.read_csv(CALC_PATH)

players.columns = players.columns.str.strip()
calc.columns = calc.columns.str.strip()

# Standardize IDs
players["ID Number"] = players["ID Number"].astype(str).str.strip()
calc["fide_id"] = calc["fide_id"].astype(str).str.strip()

print("Players shape:", players.shape)
print("Calculation records shape:", calc.shape)
print("Unique players in sample:", players["ID Number"].nunique())
print("Unique players in calculation data:", calc["fide_id"].nunique())
```

```python
# ============================================================
# Aggregate FIDE calculation-page records to player level
# ============================================================

calc_features = calc.copy()

# Convert possible numeric columns
numeric_candidates = [
    "opponent_rating",
    "rating",
    "score",
    "event_rc",
    "rc",
    "chg",
    "k_chg",
    "game_chg",
    "rating_change",
    "k_factor"
]

for col in numeric_candidates:
    if col in calc_features.columns:
        calc_features[col] = (
            calc_features[col]
            .astype(str)
            .str.replace("*", "", regex=False)
            .str.replace(",", "", regex=False)
        )
        calc_features[col] = pd.to_numeric(calc_features[col], errors="coerce")

# Detect usable columns
opponent_rating_col = "opponent_rating" if "opponent_rating" in calc_features.columns else None
if opponent_rating_col is None and "rating" in calc_features.columns:
    opponent_rating_col = "rating"

score_col = "score" if "score" in calc_features.columns else None

k_chg_col = None
for possible_col in ["k_chg", "chg", "rating_change", "game_chg"]:
    if possible_col in calc_features.columns:
        k_chg_col = possible_col
        break

print("Using opponent rating column:", opponent_rating_col)
print("Using score column:", score_col)
print("Using rating-change column:", k_chg_col)

# Build aggregation dictionary
agg_dict = {
    "calc_games_12m": ("fide_id", "count")
}

if "tournament_name" in calc_features.columns:
    agg_dict["calc_tournaments_12m"] = ("tournament_name", "nunique")

if opponent_rating_col is not None:
    agg_dict["avg_opponent_rating_12m"] = (opponent_rating_col, "mean")
    agg_dict["max_opponent_rating_12m"] = (opponent_rating_col, "max")
    agg_dict["games_vs_2000plus_12m"] = (opponent_rating_col, lambda x: (x >= 2000).sum())

if score_col is not None:
    agg_dict["total_score_12m"] = (score_col, "sum")

if k_chg_col is not None:
    agg_dict["total_k_chg_12m"] = (k_chg_col, "sum")
    agg_dict["avg_game_chg_12m"] = (k_chg_col, "mean")

calc_player_features = (
    calc_features
    .groupby("fide_id")
    .agg(**agg_dict)
    .reset_index()
)

# Score percentage
if "total_score_12m" in calc_player_features.columns:
    calc_player_features["score_percentage_12m"] = (
        calc_player_features["total_score_12m"] / calc_player_features["calc_games_12m"]
    )

print("Player-level calculation features shape:", calc_player_features.shape)
display(calc_player_features.head())
```

```python
# ============================================================
# Merge player sample with calculation-page features
# ============================================================

model_df = players.merge(
    calc_player_features,
    left_on="ID Number",
    right_on="fide_id",
    how="left"
)

# Flag calculation-page availability
model_df["has_calculation_data"] = model_df["calc_games_12m"].notna().astype(int)

# Fill count-based enrichment variables with 0
count_cols = [
    "calc_games_12m",
    "calc_tournaments_12m",
    "games_vs_2000plus_12m"
]

for col in count_cols:
    if col in model_df.columns:
        model_df[col] = model_df[col].fillna(0)

print("Final model_df shape:", model_df.shape)
print("\nCalculation data coverage:")
print(model_df["has_calculation_data"].value_counts(dropna=False))

display(model_df.head())
```

```python
# ============================================================
# Save final modelling dataset
# ============================================================

FINAL_MODEL_PATH = PROCESSED_DIR / "player_features_final.csv"

model_df.to_csv(FINAL_MODEL_PATH, index=False)

print("Saved final modelling dataset to:")
print(FINAL_MODEL_PATH)
print("Final shape:", model_df.shape)
```

RQ1 hypothesis testing

```python
# ============================================================
# Step 3: RQ1 Hypothesis Test
# Current rating only vs multivariate baseline model
# ============================================================

import statsmodels.api as sm
from statsmodels.stats.anova import anova_lm

rq1_features = [
    "rating_start",
    "total_games_12m",
    "active_months_12m",
    "inactive_months_12m"
]

rq1_data = model_df[[target_col] + rq1_features].dropna().copy()

print("RQ1 sample size:", rq1_data.shape)

y = rq1_data[target_col]

# Null model: rating only
X_null = sm.add_constant(rq1_data[["rating_start"]])
model_null = sm.OLS(y, X_null).fit()

# Full model: rating + activity features
X_full = sm.add_constant(rq1_data[rq1_features])
model_full = sm.OLS(y, X_full).fit()

anova_rq1 = anova_lm(model_null, model_full)

print("RQ1 Null Model R-squared:", round(model_null.rsquared, 4))
print("RQ1 Full Model R-squared:", round(model_full.rsquared, 4))
print("RQ1 Full Model Adjusted R-squared:", round(model_full.rsquared_adj, 4))

print("\nRQ1 Nested Regression F-test:")
display(anova_rq1)

print("\nRQ1 Full Model Summary:")
print(model_full.summary())
```

```python
# ============================================================
# Save RQ1 outputs
# ============================================================

anova_rq1.to_csv(OUTPUT_TABLES_DIR / "rq1_nested_regression_test.csv")

rq1_summary = pd.DataFrame({
    "Metric": [
        "Null Model R-squared",
        "Full Model R-squared",
        "Full Model Adjusted R-squared",
        "F-statistic",
        "p-value"
    ],
    "Value": [
        model_null.rsquared,
        model_full.rsquared,
        model_full.rsquared_adj,
        anova_rq1.loc[1, "F"],
        anova_rq1.loc[1, "Pr(>F)"]
    ]
})

rq1_summary.to_csv(OUTPUT_TABLES_DIR / "rq1_summary.csv", index=False)
display(rq1_summary)
```

regression model comparison

```python
# ============================================================
# Regression model comparison for 12-month rating gain
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
import pandas as pd

target_col = "rating_gain_12m"

baseline_features = [
    "rating_start",
    "total_games_12m",
    "active_months_12m",
    "inactive_months_12m"
]

enriched_features = [
    "calc_games_12m",
    "calc_tournaments_12m",
    "avg_opponent_rating_12m",
    "score_percentage_12m",
    "games_vs_2000plus_12m",
    "total_k_chg_12m"
]

categorical_candidates = ["Sex", "Title", "rating_band"]
categorical_features = [col for col in categorical_candidates if col in model_df.columns]

regression_features = baseline_features + enriched_features + categorical_features

# Use enriched subset for enriched model comparison
reg_data = model_df[model_df["has_calculation_data"] == 1].copy()
reg_data = reg_data[[target_col] + regression_features].dropna(subset=[target_col]).copy()

X = reg_data[regression_features]
y = reg_data[target_col]

numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X.select_dtypes(exclude=[np.number]).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.01, max_iter=10000),
    "Elastic Net": ElasticNet(alpha=0.01, l1_ratio=0.5, max_iter=10000),
    "Random Forest": RandomForestRegressor(
        n_estimators=300,
        random_state=42,
        min_samples_leaf=5
    ),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

regression_results = []

for model_name, model in models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)

    # Compatible RMSE calculation
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)

    r2 = r2_score(y_test, y_pred)

    regression_results.append({
        "Model": model_name,
        "MAE": round(mae, 2),
        "RMSE": round(rmse, 2),
        "R2": round(r2, 3)
    })

regression_results_df = (
    pd.DataFrame(regression_results)
    .sort_values("RMSE")
    .reset_index(drop=True)
)

display(regression_results_df)

# Save results
regression_results_df.to_csv(
    OUTPUT_TABLES_DIR / "regression_model_results.csv",
    index=False
)

print("Saved:", OUTPUT_TABLES_DIR / "regression_model_results.csv")
```

This means the Random Forest model explains about 39.0% of the variation in 12-month rating gain on the test set. Its average absolute error is about 38.6 rating points, and its RMSE is about 61.3 rating points.

The tree-based models outperform linear models, which suggests that rating gain is not purely linear. Interactions such as rating level × activity, activity × opponent strength, or score percentage × opposition strength may matter.

Regression models were trained to predict 12-month Standard rating gain using the enriched feature set. Among the tested models, Random Forest achieved the best performance, with MAE = 38.61, RMSE = 61.34, and R² = .390. Gradient Boosting was the second-best model, with MAE = 41.17, RMSE = 64.14, and R² = .334. Linear and regularized regression models performed lower, with R² values between .311 and .323. These results suggest that non-linear relationships and feature interactions are important in predicting youth chess rating gain. The Random Forest model is therefore the strongest candidate regression model at this stage, although linear models remain useful for interpretability and hypothesis testing.

```python
# ============================================================
# Classification model for crossing 1800 within 12 months
# ============================================================

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, average_precision_score, confusion_matrix,
    classification_report
)
import pandas as pd
import numpy as np

class_target = "cross_1800_12m"

classification_features = [
    "rating_start",
    "total_games_12m",
    "active_months_12m",
    "inactive_months_12m",
    "calc_games_12m",
    "calc_tournaments_12m",
    "avg_opponent_rating_12m",
    "score_percentage_12m",
    "games_vs_2000plus_12m",
    "total_k_chg_12m",
    "Sex",
    "Title",
    "rating_band"
]

classification_features = [col for col in classification_features if col in model_df.columns]

class_df = model_df[[class_target] + classification_features].dropna(subset=[class_target]).copy()

X = class_df[classification_features]
y = class_df[class_target]

numeric_features = X.select_dtypes(include=[np.number]).columns.tolist()
categorical_features = X.select_dtypes(exclude=[np.number]).columns.tolist()

numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(
    transformers=[
        ("num", numeric_transformer, numeric_features),
        ("cat", categorical_transformer, categorical_features)
    ]
)

class_models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, class_weight="balanced"),
    "Random Forest Classifier": RandomForestClassifier(
        n_estimators=300,
        random_state=42,
        class_weight="balanced",
        min_samples_leaf=3
    ),
    "Gradient Boosting Classifier": GradientBoostingClassifier(random_state=42)
}

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)

classification_results = []
confusion_outputs = {}

for model_name, model in class_models.items():
    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])
    
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)
    
    if hasattr(pipe.named_steps["model"], "predict_proba"):
        y_prob = pipe.predict_proba(X_test)[:, 1]
    else:
        y_prob = None
    
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0)
    recall = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)
    
    roc_auc = roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan
    pr_auc = average_precision_score(y_test, y_prob) if y_prob is not None else np.nan
    
    cm = confusion_matrix(y_test, y_pred)
    confusion_outputs[model_name] = cm
    
    classification_results.append({
        "Model": model_name,
        "Accuracy": round(accuracy, 3),
        "Precision": round(precision, 3),
        "Recall": round(recall, 3),
        "F1": round(f1, 3),
        "ROC_AUC": round(roc_auc, 3),
        "PR_AUC": round(pr_auc, 3)
    })

classification_1800_results_df = (
    pd.DataFrame(classification_results)
    .sort_values("PR_AUC", ascending=False)
    .reset_index(drop=True)
)

display(classification_1800_results_df)

print("\nConfusion matrices:")
for model_name, cm in confusion_outputs.items():
    print("\n", model_name)
    print(cm)

# Save results
classification_1800_results_df.to_csv(
    OUTPUT_TABLES_DIR / "classification_cross_1800_results.csv",
    index=False
)

print("Saved:", OUTPUT_TABLES_DIR / "classification_cross_1800_results.csv")
```

It does not imply low ability; rather, it identifies players whose lack of rated activity limits observable development and may reduce future rating progress. |


These trajectory labels are operational definitions created for this study. They are not official FIDE categories and should not be interpreted as permanent judgments about a player’s ability. The labels are intended to support modelling, subgroup comparison, and practical interpretation of different development patterns.

The trajectory cutoffs were defined as operational thresholds based on the observed distribution of 12-month rating gain and practical interpretability. Because the median gain was 0 and the 75th percentile was approximately 12 rating points, small rating changes were treated as normal fluctuation. A ±20 point band was therefore used to define plateau, while gains of 20–74 points were treated as stable progress. Gains of 75 or more points were classified as high-growth because they represent a substantially larger improvement than the typical player in the sample. Players with 10 or more inactive months and no positive gain were labelled inactivity risk to capture low activity and limited observable development. These thresholds are project-defined and should be validated through sensitivity analysis in the final model.

```python
# ============================================================
# Create developmental trajectory labels for RQ4
# ============================================================

def assign_trajectory(row):
    gain = row["rating_gain_12m"]
    inactive = row.get("inactive_months_12m", np.nan)
    
    if pd.isna(gain):
        return np.nan
    
    if inactive >= 10 and gain <= 0:
        return "Inactivity risk"
    elif gain >= 75:
        return "High-growth"
    elif gain >= 20:
        return "Stable progress"
    elif -20 <= gain < 20:
        return "Plateau"
    else:
        return "Decline"

model_df["trajectory_label"] = model_df.apply(assign_trajectory, axis=1)

print("Trajectory label distribution:")
print(model_df["trajectory_label"].value_counts(dropna=False))
```
